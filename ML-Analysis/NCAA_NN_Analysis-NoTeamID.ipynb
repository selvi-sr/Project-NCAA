```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
import warnings
warnings.simplefilter('ignore')
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from yellowbrick.datasets import load_concrete
from yellowbrick.regressor import ResidualsPlot
```

```python
history_data = pd.read_csv(os.path.join('Stage2_NCAA_2013T02018.csv'))
history_data.head()
```

```python
# Assign X (data) and y (target)
X = history_data.drop(["outcome", "TeamID","NumOT"], axis=1)
y = history_data["outcome"]
print(X.shape, y.shape)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=36, activation='relu', input_dim=14))
model.add(Dense(units=36, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
 model.fit(
    X_train_scaled,
    y_train,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
deep_model = Sequential()
deep_model.add(Dense(units=6, activation='relu', input_dim=14))
deep_model.add(Dense(units=6, activation='relu'))
deep_model.add(Dense(units=2, activation='softmax'))
```

```python
deep_model.summary()
```

```python
deep_model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
deep_model.fit(
    X_train_scaled,
    y_train,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
model_loss, model_accuracy = deep_model.evaluate(
    X_test_scaled, y_test, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
Edeep_model = Sequential()
Edeep_model.add(Dense(units=6, activation='relu', input_dim=14))
Edeep_model.add(Dense(units=6, activation='relu'))
Edeep_model.add(Dense(units=6, activation='relu'))
Edeep_model.add(Dense(units=2, activation='softmax'))
```

```python
Edeep_model.summary()
```

```python
Edeep_model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
Edeep_model.fit(
    X_train_scaled,
    y_train,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = Edeep_model.evaluate(
    X_test_scaled, y_test, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
from sklearn.metrics import classification_report
```

```python
encoded_predictions = Edeep_model.predict_classes(X_test_scaled)
encoded_predictions
```

```python
target_names = ['Loss', 'Won']
```

```python
print(classification_report(y_test, encoded_predictions, target_names=target_names))
```
